```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns

data = pd.read_csv("../data/raw/Online_Retail.csv",sep=";", encoding="latin-1")

print("header of the data: ")
data.head()

print("information about the data: ")
data.info()

print("Column names: ")
data.columns

print("rows and columns: ")
data.shape
```

 # LIMPIEZA DEL DATASET

```python
data.head()
#Renombrando las columnas
data.columns = ['InvoiceNo', 'StockCode', 'Description', 'Quantity', 'InvoiceDate', 'UnitPrice', 'CustomerID', 'Country']
```

```python
print("Number of rows: ", data.shape[0])
print("Number of columns: ", data.shape[1])
print("information about the data: ")
data.info()
print("Number of null values in each column: ")
data.isnull().sum()
```

```python
#Eliminar los registros con valores nulos en la columna "CustomerID" y "Description"
print("Number of rows before removing null values: ", data.shape[0])
data = data.dropna(subset=["CustomerID", "Description"])
print("Number of rows after removing null values: ", data.shape[0])
print("Number of null values in each column after removing null values: ")
data.isnull().sum()
```

```python
#Eliminar valores duplicados en el dataset
print("Number of rows before removing duplicates: ", data.shape[0])
data = data.drop_duplicates()
print("Number of rows after removing duplicates: ", data.shape[0])

data.shape
```

```python
#Quitar los registros con valores negativos en la columna "Quantity" y "UnitPrice"
#Pasar la columna "UnitPrice" y "Quantity" a tipo numérico y corregir los valores que tienen comas en lugar de puntos decimales

data["UnitPrice"] = data["UnitPrice"].astype(str).str.replace(",", ".")
data["Quantity"] = data["Quantity"].astype(str).str.replace(",", ".")
data["UnitPrice"] = pd.to_numeric(data["UnitPrice"], errors="coerce")
data["Quantity"] = pd.to_numeric(data["Quantity"], errors="coerce")

print("Number of rows before removing negative values: ", data.shape[0])
data = data[(data["Quantity"] > 0) & (data["UnitPrice"] > 0)]
print("Number of rows after removing negative values: ", data.shape[0])

#crear una nueva columna "TotalPrice" que sea el resultado de multiplicar la columna "Quantity" por la columna "UnitPrice"
data["TotalPrice"] = data["Quantity"] * data["UnitPrice"]
data.info()
```

```python
#convertir y limpiar de cualqier dato anomalo la columna "InvoiceDate" a formato de fecha dayfirst
data["InvoiceDate"] = pd.to_datetime(data["InvoiceDate"], errors="coerce", dayfirst=True)
print(type(data["InvoiceDate"].iloc[0]))
print("Number of null values in the column 'InvoiceDate' after conversion: ", data["InvoiceDate"].isnull().sum())

```
